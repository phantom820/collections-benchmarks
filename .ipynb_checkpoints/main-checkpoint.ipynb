```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import glob
import seaborn as sns
sns.set_theme(style="whitegrid")
```

```python
COLUMNS = ['Name','Iterations','ns/op','B/op','allocs/op']
```

```python
def loadFile(fname):
    df = pd.read_csv(fname,names=COLUMNS,skiprows=4,sep='\t')
    df = df.iloc[:-2,:]
    df['Operation'] = df['Name'].str.split('-').str[-1].str.strip()
    df['Collection'] = df['Name'].apply(lambda x : x.split('/')[-1].split('-')[0])
    df = df.iloc[:,1:]
    df = df[df.columns[::-1]]
    def clean(columns):
        for column in columns:
            df[column] = df[column].str.split().str[0]
            df[column] = pd.to_numeric(df[column])
    clean(['ns/op','B/op','allocs/op'])
    return df

def loadFiles(path):
    data = []
    fnames = glob.glob(path)
    for fname in fnames:
        data.append(loadFile(fname))
    return pd.concat(data)
```

```python
lists = loadFile('lists.csv')
sets = loadFile('sets.csv')
sets_lists = loadFiles('*.csv')
```

### Lists

```python
ax = plt.figure(figsize=(10,10))
lists.groupby(['Collection','Operation'])['ns/op'].mean().unstack().plot.bar(subplots=True,ax=ax,width=.4)
ax.tight_layout()
```

### Sets

```python
ax = plt.figure(figsize=(15,10))
sets.groupby(['Collection','Operation'])['ns/op'].mean().unstack().plot.bar(subplots=True,ax=ax,width=.4)
ax.tight_layout()
```
