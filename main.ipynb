```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import glob
import seaborn as sns
sns.set_theme(style="whitegrid")
```

```python
COLUMNS = ['Collection','Operation','Time(ns)']
```

```python
def loadFile(fname):
    df = pd.read_csv(fname,names=COLUMNS)
    df['Time(ms)']  = df['Time(ns)']*1e-6
    return df

def loadFiles(path):
    data = []
    fnames = glob.glob(path)
    for fname in fnames:
        data.append(loadFile(fname))
    return pd.concat(data)
```

```python
lists = loadFile('lists.csv')
sets = loadFile('sets.csv')
queues = loadFile('queues.csv')
sets_lists = pd.concat([lists,sets])
lists
```

### Lists

```python
ax = plt.figure(figsize=(15,12))
lists.groupby(['Collection','Operation'])['Time(ms)'].mean().unstack().plot.barh(subplots=True,ax=ax,width=.3)
ax.tight_layout()
```

### Sets

```python
ax = plt.figure(figsize=(15,10))
sets.groupby(['Collection','Operation'])['Time(ms)'].mean().unstack().plot.barh(subplots=True,ax=ax,width=.3)
ax.tight_layout()
```

### Lists & Sets

```python
ax = plt.figure(figsize=(15,20))
sets_lists.groupby(['Collection','Operation'])['Time(ms)'].mean().unstack().plot.barh(subplots=True,ax=ax,width=.3)
ax.tight_layout()
```

### Queues

```python
ax = plt.figure(figsize=(15,10))
queues.groupby(['Collection','Operation'])['Time(ms)'].mean().unstack().plot.barh(subplots=True,ax=ax,width=.3)
ax.tight_layout()
```
